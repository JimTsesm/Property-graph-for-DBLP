```python
jv = data[['article','author','crossref','journal','key','title','volume','year']]
jv = jv.dropna()
jv.head(5)
```

```python
import pandas as pd
import numpy as np

FILE_PATH = 'C:/Users/webma/Desktop/BDMA/UPC/Semantic Data Management/Labs/Lab1/Dataset/bulk_csv/'
data = pd.read_csv(FILE_PATH+'author(paper).csv')
#header = pd.read_csv(FILE_PATH+'output_article_header.csv', sep=';')
#data.columns = [x[:x.find(':')] for x in header.columns]
data.reset_index()
data.head(10)
```

```python
d={}
p={}
for i,r in data.iterrows():
    if r['author'] not in d:
        d[r['author']]=[r['key']]
    else:
        d[r['author']]=d[r['author']]+[r['key']]
    p[r['key']]=0
```

```python
d['Amit Golander']
```

```python
len(data['key'].unique())
```

```python
authors = data['author'].unique()
papers = data['key'].unique()
aut=[]
pap=[]
reviewers = np.random.choice(authors,1000, replace=False)
for vr in reviewers:
    my_reviews = np.random.choice(papers,20, replace=False)
    cnt=0
    for rvs in my_reviews:
        if rvs not in d[vr] and p[rvs] < 3:
            aut=aut+[vr]
            pap=pap+[rvs]
            p[rvs]=p[rvs]+1
            cnt=cnt+1
            if cnt==3:
                break
```
