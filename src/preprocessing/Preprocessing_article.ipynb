```python
import pandas as pd
import numpy as np

FILE_PATH = 'C:/Users/webma/Desktop/BDMA/UPC/Semantic Data Management/Labs/Lab1/Dataset/'
```

```python
data = pd.read_csv(FILE_PATH+'output_article.csv', sep=';', header=None)
header = pd.read_csv(FILE_PATH+'output_article_header.csv', sep=';')
data.columns = [x[:x.find(':')] for x in header.columns]
data.head(10)
```

```python
jv = data[['article','author','crossref','journal','key','title','volume','year']]
jv = jv.dropna()
jv.head(50)
```

```python
#Journal
journal=jv['journal'].unique()
with open(FILE_PATH+'bulk_csv/journal.csv', 'a') as the_file:
    for j in journal:
        the_file.write(j+'\n')
```

```python
#Volume
volume=jv[['crossref','volume', 'year', 'journal']]
volume.columns = ['key', 'number', 'year', 'journal']
volume['key'].replace(regex=True,inplace=True,to_replace=r'journals/',value=r'')
volume['year']=volume['year'].astype(int)
volume['number']=volume['number'].astype(int)
volume.head(10)
volume.to_csv(FILE_PATH+'bulk_csv/volume(journal).csv', index=False)
```

```python
#Paper
paper = jv[['key', 'title', 'year', 'crossref']]
paper['key'].replace(regex=True,inplace=True,to_replace=r'journals/',value=r'')
paper['crossref'].replace(regex=True,inplace=True,to_replace=r'journals/',value=r'')
paper['year'] = paper['year'].astype(int)
paper.to_csv(FILE_PATH+'bulk_csv/paper(volume).csv', index=False)
paper.head(10)
```

```python
#Author
author = jv[['key','author']]
author['key'].replace(regex=True,inplace=True,to_replace=r'journals/',value=r'')
new_author = pd.DataFrame(author.author.str.split('|').tolist(), index=author.key).stack()
new_author = new_author.reset_index([0, 'key'])
new_author.columns = ['key', 'author']
new_author.to_csv(FILE_PATH+'bulk_csv/author(paper).csv', index=False)
```
